```python
from collections import defaultdict

names = ["Alob", "Bice"]
NUM_USERS = 2
d = defaultdict(list)

# os.environ['IMGUR_KEY'] = "d07ffaf89568bde3493d49d83ea3f0677f64c912"
os.environ["IMGBB_KEY"] = "7967a398737e6be8e1d9a1090d2ba7e0"

client = imgbbpy.SyncClient(os.environ["IMGBB_KEY"])


for i in range(10):
  capidea = requests.post(url, json=payload, headers=headers).json()["output"]["choices"][0]["text"]
  time.sleep(1)
  print(f"caption {i+1} idea: {capidea}")
  for user in range(NUM_USERS):
    # captionsfile = open(f"captions-user{user}.txt", "w")
    caption = requests.post(url, json={
          "model": "meta-llama/Llama-2-70b-chat-hf",
          "prompt": f"Rephrase this and change some phrasing details: {capidea}. Output only the rephrased sentence.",
          "max_tokens": 256,
          "stop": ".",
          "temperature": 0.7,
          "top_p": 0.75,
          "top_k": 50,
          "repetition_penalty": 100
      }, headers=headers).json()["output"]["choices"][0]["text"]
    print(f"caption {i+1} spinoff {user+1}: {caption}")
    time.sleep(1)
    # captionsfile.write(f"caption_{i}: {caption}\n")
    response = requests.post(url, json={
        "model": "runwayml/stable-diffusion-v1-5",
        "prompt": caption,
        "n": 4,
        "steps": 20
    }, headers={
        "Authorization": f"Bearer {os.environ['TOGETHER_API_KEY']}",
        "User-Agent": "ImageGen"
    }).json()
    print(response)
    image = response["output"]["choices"][0]
    with open(f"image_{i}.jpg", "wb") as f:
      f.write(base64.b64decode(image["image_base64"]))

    # result = requests.post(url="https://api.imgur.com/3/image",
    #             json={
    #                 'image': image["image_base64"],
    #                 'type': 'base64',
    #                 'name': 'image_3.jpg',
    #                 'title': 'image 3'
    #             },
    #             headers={
    #                 "Authorization" : f"Bearer d07ffaf89568bde3493d49d83ea3f0677f64c912"
    #             })
    # result = requests.post(url="https://api.imgbb.com/1/upload", json={
    #     "image" : image["image_base64"],
    #     "name" : f"user{user}image{i}.jpg"
    # }, headers = {
    #     "key" : os.environ["IMGBB_KEY"]
    # })
    client = imgbbpy.SyncClient(os.environ["IMGBB_KEY"])
    image = client.upload(file=f'./image_{i}.jpg')
    # print(image.url)

    # print(result.json())
    d["link"].append(image.url)
    d["description"].append(caption)
    d["user"].append(names[user])
  # captionsfile.close()

client.close()
```
